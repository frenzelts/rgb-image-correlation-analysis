## Correlation Analysis for RGB image

Horizontal, vertical, and diagonal correlation analysis of RGB image using scatter plot and correlation coefficient

```python
import cv2 
import os
from matplotlib import pyplot as plt
import numpy as np
from statistics import mean 
```

#### Read image

```python
filepath = os.path.join('.', 'ISIC_0024306.jpg')
filename = os.path.basename(filepath).split('.')[0]
im = cv2.imread(filepath)

[height, width, channel] = im.shape
```

### Horizontal Correlation

**Respective channels**

```python
color = ('Blue','Green','Red')

for i,col in enumerate(color):
    x = im[:,:-1,i]
    y = im[:,1:,i]

    plt.scatter(x, y, s=0.01, alpha=0.5, color=col)
    plt.title(col+' horizontal pixel correlation')
    plt.xlabel('Pixel value at (x,y)')
    plt.ylabel('Pixel value at (x+1,y)')
    plt.show()

    print(col+' horizontal correlation coefficient: ', np.corrcoef(x.flat, y.flat)[0,1])
```

**All channels**

```python
coeff = []

x0 = im[:,:-1,0]
y0 = im[:,1:,0]
plt.scatter(x0, y0, s=0.01, alpha=0.3, color='b')
coeff.append(np.corrcoef(x0.flat, y0.flat)[0,1])

x1 = im[:,:-1,1]
y1 = im[:,1:,1]
plt.scatter(x1, y1, s=0.01, alpha=0.3, color='g')
coeff.append(np.corrcoef(x1.flat, y1.flat)[0,1])

x2 = im[:,:-1,2]
y2 = im[:,1:,2]
plt.scatter(x2, y2, s=0.01, alpha=0.3, color='r')
coeff.append(np.corrcoef(x2.flat, y2.flat)[0,1])

plt.title('Horizontal pixel correlation')
plt.xlabel('Pixel value at (x,y)')
plt.ylabel('Pixel value at (x+1,y)')
plt.show()

print('Horizontal correlation coefficient: ', mean(coeff))
```

### Vertical Correlation

**Respective channels**

```python
color = ('Blue','Green','Red')

for i,col in enumerate(color):
    x = im[:-1,:,i]
    y = im[1:,:,i]

    plt.scatter(x, y, s=0.01, alpha=0.5, color=col)
    plt.title(col+' vertical pixel correlation')
    plt.xlabel('Pixel value at (x,y)')
    plt.ylabel('Pixel value at (x,y+1)')
    plt.show()

    print(col+' vertical correlation coefficient: ', np.corrcoef(x.flat, y.flat)[0,1])
```

**All channels**

```python
coeff = []

x0 = im[:-1,:,0]
y0 = im[1:,:,0]
plt.scatter(x0, y0, s=0.01, alpha=0.3, color='b')
coeff.append(np.corrcoef(x0.flat, y0.flat)[0,1])

x1 = im[:-1,:,1]
y1 = im[1:,:,1]
plt.scatter(x1, y1, s=0.01, alpha=0.3, color='g')
coeff.append(np.corrcoef(x1.flat, y1.flat)[0,1])

x2 = im[:-1,:,2]
y2 = im[1:,:,2]
plt.scatter(x2, y2, s=0.01, alpha=0.3, color='r')
coeff.append(np.corrcoef(x2.flat, y2.flat)[0,1])

plt.title('Vertical pixel correlation')
plt.xlabel('Pixel value at (x,y)')
plt.ylabel('Pixel value at (x,y+1)')
plt.show()

print('Vertical correlation coefficient: ', mean(coeff))
```

### Diagonal Correlation

**Respective channels**

```python
color = ('Blue','Green','Red')

for i,col in enumerate(color):
    x = im[:-1,:-1,i]
    y = im[1:,1:,i]

    plt.scatter(x, y, s=0.01, alpha=0.5, color=col)
    plt.title(col+' diagonal pixel correlation')
    plt.xlabel('Pixel value at (x,y)')
    plt.ylabel('Pixel value at (x+1,y+1)')
    plt.show()

    print(col+' diagonal correlation coefficient: ', np.corrcoef(x.flat, y.flat)[0,1])
```

**All channels**

```python
coeff = []

x0 = im[:-1,:-1,0]
y0 = im[1:,1:,0]
plt.scatter(x0, y0, s=0.01, alpha=0.3, color='b')
coeff.append(np.corrcoef(x0.flat, y0.flat)[0,1])

x1 = im[:-1,:-1,1]
y1 = im[1:,1:,1]
plt.scatter(x1, y1, s=0.01, alpha=0.3, color='g')
coeff.append(np.corrcoef(x1.flat, y1.flat)[0,1])

x2 = im[:-1,:-1,2]
y2 = im[1:,1:,2]
plt.scatter(x2, y2, s=0.01, alpha=0.3, color='r')
coeff.append(np.corrcoef(x2.flat, y2.flat)[0,1])

plt.title('Diagonal pixel correlation')
plt.xlabel('Pixel value at (x,y)')
plt.ylabel('Pixel value at (x+1,y+1)')
plt.show()

print('Diagonal correlation coefficient: ', mean(coeff))
```
